# **Uniformally Distributed Discrete Random Variable**

meaning all the outcomes are discrete (e.g. an integer) and are equally likely (the probability of each outcome is same).

*Example*: Rolling a die. Outcomes are discrete (1, 2, 3, 4, 5, 6) and all outcomes are equally likely ( here the probability of each outcome being 1/6).

````python
import numpy as np

num_throws = 10000 # number of times a die is thrown
outcomes = np.zeros(num_throws) # total outcomes of a throw of a die

for i in range(num_throws):
    outcomes[i] = np.random.randint(1,7)
    
val, cnt = np.unique(outcomes, return_counts=True) # to find out the unique values in the outcomes array and their count
prob = cnt/len(outcomes) # to find out the probability of each outcome
````

````python
import matplotlib.pyplot as plt
plt.bar(val, prob)  # creating bar graph
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Throwing a Die 10000 times and observing the results')
plt.show()
plt.close();
````

## **Cumulative Mass Function of a Uniformly Distributed Random Variable**

````python
plt.plot(val, np.cumsum(prob)) # plotting line graph 
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Mass Function of a Uniformly Distributed Random Variable')
plt.show()
plt.close();
````

### **Plotting PMF and CMF of a Uniform Random Variable side by side** 

````python
plt.figure()
plt.subplot(211)
plt.bar(val, prob)  # creating bar graph


plt.subplot(212)
plt.plot(val, np.cumsum(prob)); # plotting line graph

````

# **Non-Uniform Discrete Random Variable** 

meaning each outcome is discrete however, certain outcomes are more likely than others.

*Example:* Rolling a biased die. 

* Probability of 1: ```0.27```
* Probability of 2: ```0.23```
* Probability of 3: ```0.15```
* Probability of 4: ```0.05```
* Probability of 5: ```0.20```
* Probability of 6: ```0.10```

````python
num_throws = 10000 # number of times a die is thrown
biased_outcomes = np.zeros(num_throws) # creating an array where outcomes of the biased coin are stored

for i in range(num_throws):
  biased_outcomes[i] = np.random.choice(['1', '2', '3', '4', '5', '6'], p=[0.27, 0.23, 0.15, 0.05, 0.2, 0.1])

values, count = np.unique(biased_outcomes, return_counts=True) # to find out the unique values in the outcomes array and their count

biased_probability = count/len(biased_outcomes) # to find out the probability of each outcome
````

````python
plt.bar(values, biased_probability)  # creating bar graph
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Throwing a Biased Die 10000 times and observing the results')
plt.show()
plt.close();
````

## **Cumulative Mass Function of a Uniformly Distributed Random Variable**

````python
plt.plot(val, np.cumsum(biased_probability))
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Mass Function of a Non-Uniform Discrete Random Variable')
plt.show()
plt.close();
````

### **Plotting PMF and CMF of a Uniform Random Variable side by side** 

````python
plt.figure()
plt.subplot(211)
plt.bar(values, biased_probability)  # creating bar graph


plt.subplot(212)
plt.plot(val, np.cumsum(biased_probability)); # plotting line graph
````
